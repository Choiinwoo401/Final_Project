구별 침수 빈도

```python

flood_counts = df.groupby(['자치구', '침수연도']).size().unstack(fill_value=0)

flood_counts.plot(kind='bar', figsize=(14, 8), width=0.8)
plt.title('서울시 구별 연도별 침수 횟수')
plt.xlabel('자치구')
plt.ylabel('침수 횟수')
plt.xticks(rotation=45, ha='right')
plt.legend(title='연도')
plt.tight_layout()
```

```python
# 구별, 연도별 침수 횟수를 계산합니다.
flood_counts = df.groupby(['자치구', '침수연도']).size().unstack(fill_value=0)

# 그래프 그리기
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 8))

# 첫 번째 subplot: 2022년 제외
years = flood_counts.columns
bar_width = 0.1  # 각 막대의 너비
index = np.arange(len(flood_counts))

# 첫 번째 y축 (2022년 제외)
for i, year in enumerate(years):
    if year != 2022:
        ax1.bar(index + i * bar_width, flood_counts[year], bar_width, label=str(year))

ax1.set_xlabel('자치구')
ax1.set_ylabel('침수 횟수')
ax1.set_xticks(index + bar_width * (len(years) - 1) / 2)
ax1.set_xticklabels(flood_counts.index, rotation=45, ha='right')
ax1.legend(title='연도')


# 두 번째 subplot: 2022년
ax2.bar(index, flood_counts[2022], bar_width, label='2022', color='red')
ax2.set_xlabel('자치구')
ax2.set_ylabel('침수 횟수 (2022년)')
ax2.set_xticks(index)
ax2.set_xticklabels(flood_counts.index, rotation=45, ha='right')
ax2.legend(title='연도')



```
